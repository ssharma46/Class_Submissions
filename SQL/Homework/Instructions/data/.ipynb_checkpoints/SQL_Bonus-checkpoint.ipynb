```python
##Import dependencies
import pandas as pd
import numpy as np
import sqlalchemy as sa
from sqlalchemy import create_engine
import psycopg2
import matplotlib.pyplot as plt
```

```python
#Create Engine and connection to Database
engine = create_engine('postgres://postgres:PASSWORD@localhost:5432/SQL_HW')
conn = engine.connect()
```

```python
#Query records from Titles Table
titles_df = pd.read_sql("SELECT * FROM titles", conn)
titles_df.head()
```

```python
#Query records from Salaries Table
salaries_df = pd.read_sql("SELECT * FROM salaries", conn)
salaries_df.head()
```

```python
#Merge Data
salary_titledf = pd.merge(titles_df, salaries_df, on="emp_no", how="inner")
salary_titledf.head()
```

```python
#Group data by Title
salary_titlegrp = salary_titledf.groupby("title").mean()
salary_titlegrp.head()
```

```python
#Drop emp_no column
clean_salary_title = salary_titlegrp.drop(columns = "emp_no")
clean_salary_title.head()

#Reset Index
clean_salary_title = clean_salary_title.reset_index()
clean_salary_title
```

```python
##Create Bar Chart using Matplotlib

#Set axes
x_axis = clean_salary_title["title"]
y_axis = clean_salary_title["salary"]

#Set Title and Labels and fix x-axis labels
plt.title("Average Employee Salary by Job Title")
plt.ylabel("Salaries (USD)")
plt.xlabel("Employee Titles")
plt.xticks(rotation=90)

#Create Bar Chart
plt.bar(x_axis, y_axis, align="center", alpha=.75, color=["plum", "mediumspringgreen", "slateblue", "gold", "royalblue", "lightslategray", "tomato"])
plt.show
```
